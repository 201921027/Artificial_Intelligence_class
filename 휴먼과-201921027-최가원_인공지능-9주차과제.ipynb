```python
import tensorflow as tf

batch_size= 128
num_classes= 10
epochs= 20
```

```python
(x_train, y_train), (x_test, y_test)= tf.keras.datasets.mnist.load_data()

x_train= x_train.reshape(60000, 784) # 60000개 입력이미지를 2차원에서 1차원 벡터로 변경
x_test= x_test.reshape(10000, 784) # 10000개 입력이미지를 2차원에서 1차원 벡터로 변경
```

```python
x_train= x_train.astype('float32')
x_test= x_test.astype('float32')
x_train /= 255 # 입력 이미지의 픽셀값이 0.0 ~ 1.0 사이의 값이 되게 한다
x_test /= 255
```

```python
# 클래스의 개수에 따라 클래스를 원핫 인코딩으로 출력
y_train= tf.keras.utils.to_categorical(y_train, num_classes)
y_test= tf.keras.utils.to_categorical(y_test, num_classes)
```

```python
model= tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(512, activation='sigmoid', input_shape=(784,)))
# model.add(tf.keras.layers.Dense(128, activation='sigmoid', input_shape=(512,)))
model.add(tf.keras.layers.Dense(num_classes, activation='sigmoid'))

model.summary()
```

```python
sgd= tf.keras.optimizers.SGD(lr=0.1)

# 손실함수는 제곱오차함수로, 학습알고리즘은 SGD로
model.compile(loss='mean_squared_error',
              optimizer= sgd,
              metrics=['accuracy'])

history= model.fit(x_train, y_train,
                   batch_size= batch_size,
                   epochs= epochs)

score= model.evaluate(x_test, y_test, verbose=0)
print('테스트 손실값: ', score[0])
print('테스트 정확도: ', score[1])
```
