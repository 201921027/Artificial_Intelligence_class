```python
import tensorflow as tf
import numpy as np
```

```python
from tensorflow.keras.models import Sequential
```

```python
X= np.array([[0,0],[0,1],[1,0],[1,1]])
y= np.array([[0], [1], [1], [0]])

model= Sequential() # 모델 생성
model.add(tf.keras.layers.Dense(2, input_dim=2, activation='sigmoid')) # hidden layer 1 (#노드= 2)
model.add(tf.keras.layers.Dense(1, activation='sigmoid')) # hidden layer 2 (#노드= 1)

sgd= tf.keras.optimizers.SGD(lr=0.1) # 사용하는 최적화 알고리즘: SGD(경사하강법) (학습률= 0.1)
model.compile(loss='mean_squared_error', optimizer= sgd) #사용하는 loss f= mean squared error f
```

```python
# X, y 데이터로 학습, train data 수가 적으므로 batch로 분리하지 않음. epoch=1000회 반복
model.fit(X, y, batch_size=1, epochs=20000) 
```

```python
print(model.predict(X))
```
